*Customers.csv* which lists all customers of Widgets Inc.<br>
*Orders.csv* which lists all orders placed in 2019 & 2020. <br>

**Company_ID** in  *Customers.csv*  is the same attribute as **Customer_ID** in  *Orders.csv*

- Compile a list of customers who have placed multiple orders over the course of 2019-2020 **[10 Points]**
- Identify customers who placed multiple orders in a single month **[10 Points]**

***Note: Your list should only contain the names of customers in alphabetical order***

 ***You MUST use Dataframe and Series object operations such as apply, groupby, merge, concat and masking to obtain the above lists. Anything else will receive 0 credit even if you somehow managed to get the right answer!***



```python
customer = pd.read_csv('Customers.csv')
customer.head()
```

```python
order = pd.read_csv('Orders.csv')
order
```

```python
type(order['Date'].iloc[0])
```

```python
order['Date'] = pd.to_datetime(order['Date'])
```

```python
grouped = order.groupby('Customer_ID').count()
multi_purchase = grouped[grouped['Date']>1]
multi_purchase
```

```python
final = pd.merge(multi_purchase,customer,left_index=True, right_on='Company_ID')
final
```

```python
answer_1 = final['Company_Name'].to_list()
sorted(answer_1)
```

```python
type(order['Date'].iloc[0])
```

```python
order['month'] = order['Date'].apply(lambda x: x.month)
order
```

```python
grouped_1 = order.groupby(['Customer_ID','month']).count()
grouped_1
```

```python
multi_month_purchase = grouped_1[grouped_1['Amount']>1]
multi_month_purchase
```

```python
final_2 = pd.merge(multi_month_purchase,customer,left_on='Customer_ID', right_on='Company_ID')
final_2
```

```python
answer_2 = final_2['Company_Name'].to_list()
sorted(answer_2)
```
